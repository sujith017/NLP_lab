```python
import nltk
```

```python
nltk.download('brown')
```

```python
from nltk.corpus import brown
```

```python
print(brown.categories())
```

```python
print(brown.sents(categories='humor'))
```

```python
print(brown.tagged_sents(categories='news'))
```

```python
tagged_words=brown.tagged_words(categories='fiction')
print(tagged_words)
```

```python
nouns=[(word,tag) for word,tag in tagged_words if any(noun_tag in tag for noun_tag in['NP','NN'])]
print(nouns)
```

```python
pronouns=[(word,tag) for word,tag in tagged_words if any(pronoun_tag in tag for pronoun_tag in ['PP','PRP$'])]
print(pronouns)
```

```python
pronouns_freq=nltk.FreqDist([word for word,tag in pronouns])
print(pronouns_freq.most_common(3))
```

```python
nltk.download('wordnet')
```

```python
from nltk.corpus import wordnet as wn
```

```python
word='hi'
word_synsets=wn.synsets(word)
print(word_synsets)
```

```python
print(word_synsets[0].definition())
```

```python
word_synsets=wn.synsets(word)
for synset in word_synsets:
    print('synset name',synset.name())
```

```python
word_synsets = wn.synsets(word)
for synset in word_synsets:
    print(f"\nSynset: {synset.name()}")
    # Synonyms
    synonyms = set()
    for lemma in synset.lemmas():
        synonyms.add(lemma.name())
    print(f"Synonyms: {synonyms}")
    # Antonyms
    antonyms = set()
    for lemma in synset.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())
    print(f"Antonyms: {antonyms if antonyms else 'None'}")
```

```python
word_synsets = wn.synsets(word)
for synset in word_synsets:
    print(f"\nSynset: {synset.name()}")
    hypernyms = synset.hypernyms()
    if hypernyms:
        print("Hypernyms:")
        for hypernym in hypernyms:
            print(f"  - {hypernym.name()} : {hypernym.definition()}")
    else:
        print("No hypernyms found.")
```
